```python
import pandas as pd
import numpy as np
from bs4 import BeautifulSoup
import requests
import matplotlib.pyplot as plt
from API_Keys import Alpha_API_Key
from pprint import pprint
import json
```

```python
url = "https://www.treasury.gov/resource-center/data-chart-center/interest-rates/pages/TextView.aspx?data=yieldYear&year=2021"
```

```python
response = requests.get(url)
```

```python
soup = BeautifulSoup(response.text)
```

```python
# Look at webpage HTML

print(soup.prettify())
```

```python
Table= soup.find_all('table', {"class": "t-chart"})

print(Table)
```

```python
header=Table[0].find_all('th')

print(header)
```

```python
head=[]

for each in header:
    head.append(each.text)
    
head
```

```python
# Look at rows in table

Rows= Table[0].find_all('tr')

data=[]
  
print(Rows)
```

```python
# Grab data points from each row

data=[]

for each in Rows:
    yields=each.find_all('td')
    data.append(yields)
    
data[1]
```

```python
#  Turn data into a list of dictionaries to create dateFrame

Yields=[]

for row in data[1:]:
    Dict={}
    Dict['Date']=row[0].text
    Dict['1 Month']=row[1].text
    Dict['2 Month']=row[2].text
    Dict['3 Month']=row[3].text
    Dict['6 Month']=row[4].text
    Dict['1 Year']=row[5].text
    Dict['2 Year']=row[6].text
    Dict['3 Year']=row[7].text
    Dict['5 Year']=row[8].text
    Dict['7 Year']=row[9].text
    Dict['10 Year']=row[10].text
    Dict['20 Year']=row[11].text
    Dict['30 Year']=row[12].text
    Yields.append(Dict)
    
```

```python
Yields_DF=pd.DataFrame(Yields)

Yields_DF
```

```python
type(Yields_DF['Date'][1])
```

```python
# add columns to dataframe calculating delta between durations

Yields_DF["10-1year"]=((Yields_DF['10 Year'].astype(float)) - (Yields_DF['1 Year'].astype(float)))

Yields_DF["10-5year"]=((Yields_DF['10 Year'].astype(float)) - (Yields_DF['5 Year'].astype(float)))
```

```python
# Change date from string to timestamp

Yields_DF['Date']=pd.to_datetime(Yields_DF['Date'])

Yields_DF.head()
```

```python
type(Yields_DF['Date'][1])
```

```python
plt.figure(figsize=(20,5))
plt.plot(Yields_DF['Date'], Yields_DF['10-1year'], label='10-1')
plt.plot(Yields_DF['Date'], Yields_DF['10-5year'], label='10-5')

# x_axis= np.arange(len(Yields_DF['Date']))
# tick_locations= [value for value in x_axis]
# plt.xticks(tick_locations, Yields_DF['Date'], rotation=90)

plt.legend()

plt.xlabel('Date')
plt.ylabel('Delta')
plt.show()
```

```python
url= f'https://www.alphavantage.co/query?function=TIME_SERIES_DAILY&symbol=QQQ&outputsize=full&apikey={Alpha_API_Key}'
```

```python
response= requests.get(url).json()
```

```python
pprint(response)
```

```python
type(response)
```

```python
# Turn dictionary into DataFrame

Nasdaq= pd.DataFrame.from_dict(response['Time Series (Daily)'], orient='index')
```

```python
Nasdaq
```

```python
# Reset index to make date a usable column and clean up column names

Nasdaq=Nasdaq.reset_index()
Nasdaq.rename(columns={'index': 'Date', '1. open':'open', '2. high':'high', '3. low':'low', '4. close':'close', '5. volume':'volume'}, inplace=True)
```

```python
Nasdaq
```

```python
Nasdaq['Date']=pd.to_datetime(Nasdaq['Date'])
type(Nasdaq['Date'][1])
```

```python
Nasdaq.sort_values(by=['Date'],ascending=True, inplace=True)

Nasdaq.head()
```

```python
# Reduce Nasdaq DF to 2021

Nasdaq2021=Nasdaq[(Nasdaq['Date']>='2021-01-01') & (Nasdaq['Date']<'2022-01-01')]
```
